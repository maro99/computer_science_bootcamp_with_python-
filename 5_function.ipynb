```
# include <iostram>
using namespace std;

void change_value(int *x, int value) # 1
{
    *x = value;                      # 2
    cout << "x: " << *x << "in change_value" << endl;
}

int main(void)
{
    int x = 10;                      # 3
    change_value(&x,20);              # 4
    cout << "x: " << x << "in main" << endl;
    
    return 0;
}

실행결과
x : 20 in chhange_value
x : 20 in main

분석  
int * 는 int형 포인터를 의미한다.
다음은 change value()함수 호출시 #4 *x = value 가 실행 #2 되기 직전에본 스택 프레임 모습이다.  

   change_value스택프레임  / [ x = 0x1111 1111]               
                        l  [ value = 20 ]
   main 스택프레임         \>[ x =10 ]   adress: 0x1111 1111 
 

#4 에서 &x는 변수 x가 위치한 메모리 공간의 첫번째 바이트 주소 값을 전달 한다는 의미.  

#1 의 *x는 포인터 변수를 의미.(포인터 변수는 메모리주소를저장.)  
포인터 변수가 주소값을 저장한다는 것은 위그림처럼  
change_value 스택 프레임 안에 있는 포인터 x가 화살표를 따라 main함수 지역의 지역변수 x를 가리키는 것과 같은 의미이다. 가리킨다는 = 참조 . 

이렇게 인자로 변소의 참조를 전달하는 방식을 ""참조에 의한 전달"" 이라한다.

#2에서 *x는 역참조라 하며 x에 저장된 주소값으로 접근하고, 
main함수의 지역변수 x가 있는 메모리 공간에 value값을 대입 할 수 있다.  
(함수에 참조 인자 전달해서 함수 안에서 main함수의 스택프레임의 지역변수 x값 변경 한것.) 

```

### 2.3 객체 참조에 의한 전달(파이썬)- 변경 불가능한  객체 전달시   

파이썬에선 객체참조에 의한 전달(call by object reference)라는 방식으로 인자 전달한다. 
변경불가능한 객체 즉 상수 객체를 인자로 전달해 보겠다.  

```
def change_value(x, value): #3
    x = value               #4
    print("x:{} in change_vlaue".format(x))
    
if __name__ == "__main__":
    x = 10                  #1
    change_value(x,20)      #2
    print("x : {} in main".format(x))
    
실행결과 
x : 20 in change_value
x : 10 in main

분석  
# 2의 change_value()함수 호출 되면서 인자로 #1의 x를 전달한다.   
이때 change_vluae 스택 프레임이 생성되면서 #3의 인자 x는 함수를 호출한 영역에 있는 #1의 x를 참조한다. 

    
    change_value  [ x ]_____
                            \
                  [value]\  /[10]
                          \/
                          /\
    main          [ x ]__/  \_[20]
    
    
"" 파이썬의 변수는 C언어처럼 변수라는 메모리 공간에 값을 직접 저장하지 않는다""
변수 이름 이 값 객체를 가리키는데 주목해라.  


함수 스택 프레임 안에서 x 값이 변경되었지만 함수를 호출한 쪽에서는 x값이 변경되지 않는다. 

    change_value  [ x ]___
                          \
                  [value]\ \ /[10]
                          \/\
                          /\ \
    main          [ x ]__/  \_[20]  
    
상수 객체는 변경 불가능이다. 
""변수 값을 바꾼다는 의미는 변수 이름이 가리키는 메모리 공간의 값을 직접 바꾸는 게 아니라  
바꾸고자 하는 상수 객체를 참조하는 것이다.""  
#4는 x 가 value가 가리키는 상수객체 20을 가리키게한다. 

다음은 함수 호출 완료한 다음 살펴본 스택 프레임 모습이다. 

                            /[10]
                           / 
                          / 
    main          [ x ]__/  

chanage_value 스택 프레임이 사라지면서 change_value()함수의 지역변수 x와 value 가 사라졌다. 
또한 레퍼런스 카운트가 0이된 상수 객체 20도 사라졌다.  
x를 출력하면 1-이 아온다. 값이 변경되지 않았다.(2.2의 참조에 의한 전달방식, 값에의한 전달방식과 과정 다르다.)
```

**레퍼런스 카운트** 란 ? 
어떤 객체가 몇번 참조되는지 count하는것.
위의 예를 들면 상수객체 20이 두번 참조 되다가 change_vlaue의 스텍프레임 사라지면서 참조되는 곳이 zero가 됬는데 메모리에 상수객체 20을 할당하고 있을 필요가 없어지므로 레퍼런스카운트가 0이된 상수객체 20에대한 메모리를 가비지 컬랙터가 해제한다.


### 2.4 객체 잠조에 의한 전달( 파이썬) - 변경 가능한 객체를 전달할대 

변경가능 객체인 리스트를 인자로 전달해 보겠다. 

```
def func(li):
    li[0] = 'Iam your father!'
    
if __name__ == "__main__":
    li = [1,2,3,4]
    func(li)
    print(li)

실행결과
['Iam your fahter', 2,3,4]

분석
리스트 요소에 값 할당하니 리스트의 요소가 변경되었다.
( 2.3의 변경불가능 객체 전달시 변경 안됬는데... 왜지?)
아래 경우와 비교해보겠다. 


def func(li):
    li = ['Iam your father!',2,3,4] #1
    
if __name__ == "__main__":
    li = [1,2,3,4]
    func(li)
    print(li)

실행결과
[1,2,3,4]

분석 
함수 안에서 리스트 li에 새로운 리스트를 통째로 할당한 경우 리스트가 변경되지 않았다.


두 코드의 차이는 
전자는 참조한 리스트에 접근해 변경을 시도했고 
후자는 아예 다른 리스트를 메모리 공간에 새로 만든 다음 이를 참조해 리스트를 변경 했다는것.  


전자의 경우 
                      ___[ (), (), (), () ]
                     / /    l    l   l  l
        func  [ li ]  /     |    (2) (3) (4) 
        main  [ li ] /      |
                            |
                           ('Iam your fahter!')


-> func,main안의 li둘다 같은 메모리 공간 참조한다. 
   (변경불가능 객체의 경우 새로운 객체 예를들어 10말고 20 만들어서 참조를 통해 새로 만든 객체 가리키는 방법밖 없지만) 리스트는 변경 가능 객체여서 첫번째 요소값을 변경할 때 값 객체(iamyoutfahter) 만 새로운 공간에 만들어 참조하면 된다.즉 값을 변경하기 위해 리스트 자체를 다른 메모리공간에 새로 할당할 필요가 없다. 



후자의 경우  

                      ___[ (), (), (), () ]
                     /      l    l   l  l
        func  [ li ]        |    (2) (3) (4) 
        main  [ li ]        |
                   \        |
                    \      ('Iam your fahter!')
                     \ 
                       [ (), (), (), () ]
                          l    l   l  l
                          (1)  (2) (3) (4) 
                          
-> 다른 메모리 공간에 새로운 리스트를 만들어 li로 참조를 했다.  
요소가 아니라 리스트 자체를 변경했다.  
이경우는 함수 호출이 끝나면 func 스택프레임이 사리지므로 새로 만들어진 리스트는 삭제된다.
main지역 변수인 리스트 li는 변경되지 않는다.  
```

**위에것들 요약** 
```
1. 함수 인자로 변경 불가능 객체를 전달해 값을 변경할 수 없다.
   함수 안에서 새 객체 만든후 참조 하여 바꾸려 하면 함수 호출이 끝나고 스택 프레임 사라지면서 참조도 사라짐.

2. 리스트,딕셔너리 등 변경 가능 객체도 함수 안에서 새로운 객체 만들 경우 함수 호출이 끝나면 객체 사라진다. 

3. 변경 가능 객체를 인자로 전달할 대도 인자로 전달된 객체에 접근하여 변경해야만 함수를 호출한 쪽의 객체를 변경할 수 있다. 이런 파이썬의 인자 전달 방식을 "" 객체 참조에 의한 전달방식 이라 한다.""
```

#### 래퍼런스 카운트란? 

#### 변경불가능 객체를 함수 인자로 전달하는법.

## 3 람다함수
